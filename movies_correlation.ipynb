```python
# Creando un marco de datos de columnas numéricas, debido al error 
# "could not convert string to float: 'The Shining'".
movies_numeric = movies.select_dtypes(include=['number'])

# Matriz de Correlación, usando el Método Pearson.
movies_numeric.corr(method ='pearson')
```

```python
# Copiando la matriz de correlación de columnas númericas.
correlation_matrix = movies_numeric.corr()

# Creando un Mapa de Calor a partir de la matriz de correlación de las películas.
sns.heatmap(correlation_matrix, annot=True)

# Leyenda del Gráfico.
plt.title("Correlación de Películas")
plt.xlabel("Datos en X")
plt.ylabel("Datos en Y")
plt.show()
```

```python
# Copiando una matriz de correlación usando factorize para 
# asignar un valor numérico aleatorio para cada valor categórico único.
correlation_matrix = movies.apply(lambda x: x.factorize()[0]).corr(method='pearson')

# Creando un Mapa de Calor a partir de la matriz de correlación copiada.
sns.heatmap(correlation_matrix, annot=True)

# Leyenda del Gráfico
plt.title("Correlación de Películas")
plt.xlabel("Datos en X")
plt.ylabel("Datos en Y")
plt.show()
```

```python
# Copiando una matriz de correlación usando factorize para 
# asignar un valor numérico aleatorio para cada valor categórico único.
correlation_mat = movies.apply(lambda x: x.factorize()[0]).corr()

# Obteniedo pares correlativos, convertiendo niveles de fila especificados a niveles de columna.
corr_pairs = correlation_mat.unstack()

# Imprimiendo pares correlativos.
print(corr_pairs)
```

```python
# Aplicando ordenamiento "quicksort" a los pares correlativos obtenidos.
sorted_pairs = corr_pairs.sort_values(kind="quicksort")

# Imprimiendo pares correlativos ordenados.
print(sorted_pairs)
```

```python
# Obteniendo los pares que tienen una alta correlación (> 0.5)
strong_pairs = sorted_pairs[abs(sorted_pairs) > 0.5]

# Imprimiento resultados.
print(strong_pairs)
```

```python
# Copiando el marco de datos.
movies_numerized = movies

# Creando un marco de datos que contenga unicamente columnas numericas.
for col_name in movies_numerized.columns:
    if(movies_numerized[col_name].dtype == 'object'):
        movies_numerized[col_name] = movies_numerized[col_name].astype('category')
        movies_numerized[col_name] = movies_numerized[col_name].cat.codes

# Ver las primeras y ultimas filas del nuevo marco de datos numerado.
movies_numerized
```

```python
# Matriz de Correlación entre todas las columnas númericas, usando el Método Pearson.
movies_numerized.corr(method='pearson')
```

```python
# Creando una matriz de correlación usando el método pearson 
# con el marco de datos numerado.
correlation_matrix = movies_numerized.corr(method='pearson')

# Creando un mapa de calor a partir de la matriz de correlación creada
sns.heatmap(correlation_matrix, annot=True)

# Detalles del Gráfico
plt.title("Matriz de Correlación de Las Películas")
plt.xlabel("Caracteristicas de la Película")
plt.ylabel("Caracteristicas de la Película")
plt.show()
```
